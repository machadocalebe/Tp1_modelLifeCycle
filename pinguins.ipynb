## Exercício 1: Explicação Básica
Escreva um parágrafo explicando o que é Machine Learning e por que é útil em diversas aplicações do mundo real.

---

- Machine Learning (Aprendizado de Máquina) é uma área da inteligência artificial que permite que computadores aprendam padrões e façam previsões ou decisões baseadas em dados, sem serem explicitamente programados para cada tarefa. Em vez de seguir regras fixas, os algoritmos de machine learning analisam grandes volumes de dados, identificam relações e ajustam seus modelos automaticamente para melhorar a performance com o tempo. Essa abordagem é extremamente útil em aplicações do mundo real como reconhecimento de voz e imagem, diagnósticos médicos, recomendações de produtos (como em e-commerces e serviços de streaming), detecção de fraudes financeiras e previsão de demanda em empresas. Ao permitir decisões mais inteligentes e automatizadas, o machine learning agrega valor a diversas áreas ao transformar dados em insights e ações.

## Exercício 2 Tipos de Machine Learning
Liste e explique brevemente os dois principais tipos de Machine Learning: Aprendizado Supervisionado e Não-Supervisionado.

---

###  Conclusão

1. **Aprendizado Supervisionado** (Supervised Learning):
- Neste tipo de aprendizado, o modelo é treinado com um conjunto de dados rotulado — ou seja, cada exemplo de treino contém as entradas (features) e as saídas desejadas (rótulos ou targets). O objetivo é fazer com que o modelo aprenda a associar as entradas às saídas para que possa prever corretamente os rótulos em novos dados.
    **Exemplos comuns:**

    - Classificação (ex: identificar se um e-mail é spam ou não)

    - Regressão (ex: prever o preço de uma casa com base em seus atributos)

2. **Aprendizado Não-Supervisionado** (Unsupervised Learning):
- Aqui, o modelo recebe apenas os dados de entrada, sem rótulos. O objetivo é encontrar padrões, estruturas ou agrupamentos escondidos dentro dos dados. Esse tipo de aprendizado é útil quando não se tem uma saída desejada previamente conhecida.
Exemplos comuns:

    - Agrupamento (Clustering): agrupar clientes com perfis de compra similares

    - Redução de Dimensionalidade: simplificar grandes volumes de dados mantendo as informações mais relevantes (ex: PCA)

## Exercício 4 Aprendizado Baseado em Instâncias vs Modelos
Explique a diferença entre aprendizado baseado em instâncias e aprendizado baseado em modelos, dando um exemplo de cada um.

---

### Conclusão

1. **Aprendizado Baseado em Instâncias**  
   - Esse tipo de aprendizado não constrói um modelo explícito durante o treinamento. Em vez disso, ele memoriza os dados de treino e toma decisões com base na semelhança entre as novas entradas e os exemplos armazenados.  
   - **Exemplo:** O algoritmo K-Nearest Neighbors (KNN).  
     - Ao prever uma nova amostra, ele compara com os dados existentes e retorna o valor mais comum (classificação) ou a média (regressão) dos k vizinhos mais próximos.

2. **Aprendizado Baseado em Modelos**  
   - Nesse caso, o algoritmo constrói um modelo geral a partir dos dados de treino, aprendendo padrões e regras. Esse modelo é então usado para fazer previsões em novos dados.  
   - **Exemplo:** A Regressão Linear.  
     - Ela ajusta uma equação (reta) aos dados para prever o valor da variável alvo com base nos atributos de entrada.

**Resumo:**  
- **Instância:** memoriza os dados → exemplo: KNN  
- **Modelo:** generaliza um padrão → exemplo: Regressão Linear

## Exercício 5 Desafios em Machine Learning
Enumere e descreva três desafios comuns enfrentados ao criar modelos de Machine Learning.

---

### Conclusão

1. **Overfitting (Sobreajuste)**  
   - Ocorre quando o modelo aprende demais os dados de treino, incluindo ruídos e exceções, o que reduz sua capacidade de generalizar para novos dados.  
   - **Solução comum:** Usar validação cruzada, regularização ou reduzir a complexidade do modelo.

2. **Dados Insuficientes ou Desequilibrados**  
   - Modelos precisam de quantidade e qualidade nos dados. Poucos dados ou classes desbalanceadas (ex: 95% da classe A e 5% da B) podem levar a previsões enviesadas ou fracas.  
   - **Solução comum:** Coleta de mais dados, técnicas de reamostragem (oversampling/undersampling) ou uso de métricas adequadas.

3. **Escolha e Engenharia de Features**  
   - A escolha errada de variáveis (features) ou a presença de variáveis irrelevantes pode atrapalhar a performance do modelo.  
   - **Solução comum:** Aplicar técnicas de seleção de features, normalização e criação de variáveis úteis (feature engineering).

## Exercício 7 Preparando os Dados
Com a base de dados "Pinguins", escreva um código em Python para separar os dados em conjuntos de treino e validação. Observe que iremos criar um classificador que diferencia a espécie "Adelie" das outras duas.

---

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split


print(f"Antes do dropna: {df.shape}")
df = df.dropna()
print(f"Após dropna: {df.shape}")


df['target'] = df['species'].apply(lambda x: 1 if x == 'Adelie' else 0)


features = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
X = df[features]
y = df['target']


X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)


sns.countplot(x='target', hue='target', data=df, palette='Set2', legend=False)
plt.title('Distribuição das Classes (0 = Outras, 1 = Adelie)')
plt.xlabel('Classe')
plt.ylabel('Quantidade')
plt.grid(axis='y')
plt.show()



print(f'Treinamento: {X_train.shape}, Validação: {X_val.shape}')
```

## Conclusão

Nesta etapa, vamos preparar os dados para construir um classificador binário, cujo objetivo é **distinguir a espécie "Adelie" das demais**.

### Etapas realizadas:
- Remoção de valores ausentes para evitar erros durante a modelagem.
- Criação de um **target binário**, onde `1` representa a espécie **Adelie** e `0` representa **outras espécies**.
- Seleção de **features numéricas** mais relevantes.
- Separação dos dados em dois conjuntos:
  - **Treinamento (80%)**
  - **Validação (20%)**

A próxima etapa será treinar um modelo com esses dados e avaliar sua performance.


## Exercício 8 Construção do Modelo Inicial
Utilizando a biblioteca Scikit-Learn, escreva um código em Python para construir um modelo de Machine Learning usando o algoritmo de K-Nearest Neighbors com a base de dados "Pinguins".

---

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
from sklearn.decomposition import PCA
import numpy as np

pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X_train)

pred_train = knn.predict(X_train)

plt.figure(figsize=(8, 5))
scatter = plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=pred_train, cmap='coolwarm', alpha=0.7)
plt.title('Visualização das Previsões do Modelo (Treino)')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.grid(True)
plt.show()
```

### 🔎 Construção do Modelo com KNN

- Utilizamos o algoritmo `K-Nearest Neighbors` com 5 vizinhos (`k=5`).
- O modelo foi treinado com os dados de treino preparados anteriormente.
- Utilizamos **PCA (Análise de Componentes Principais)** para reduzir os dados para 2 dimensões e visualizar as classificações do modelo.


## Exercício 9 Avaliação do Modelo
Após treinar o modelo de K-Nearest Neighbors com a base de dados "Pinguins", escreva um código para avaliar a acurácia do seu modelo nos dados de validação.

---

```python
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay

y_pred = knn.predict(X_val)

accuracy = accuracy_score(y_val, y_pred)
print(f"Acurácia na validação: {accuracy:.2%}")
```

```python

cm = confusion_matrix(y_val, y_pred, labels=[1, 0])
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Adelie", "Outras"])
disp.plot(cmap="Blues")
plt.title("Matriz de Confusão - Validação")
plt.grid(False)
plt.show()
```

### ✅ Interpretação

A matriz de confusão mostra o desempenho do modelo em prever corretamente as espécies.  
Diagonal principal = acertos.  
A acurácia é uma boa métrica inicial, mas a matriz revela onde o modelo erra, sendo mais informativa para decisões práticas.


## Exercício 10 Reflexão Crítica
Baseado nos exercícios anteriores, escreva um breve texto discutindo os desafios encontrados ao criar um modelo de Machine Learning e como cada etapa do processo é crucial para o sucesso do projeto.

---

**A construção de um modelo de Machine Learning envolve diversas etapas fundamentais, cada uma com seus próprios desafios e importância.**

### 🔹 1. Entendimento dos dados
Compreender a estrutura, as variáveis e o problema é o ponto de partida. No caso dos pinguins, foi necessário identificar quais colunas representavam as *features* e qual seria o *target*, além de decidir transformar o problema em uma classificação binária.

### 🔹 2. Preparação dos dados
Dados ausentes, categorias não numéricas e a separação em conjuntos de treino e validação são etapas críticas. Um dado mal preparado pode comprometer todo o desempenho do modelo.

### 🔹 3. Escolha e treinamento do modelo
Selecionar o algoritmo apropriado, como o KNN, e ajustar seus parâmetros iniciais é um passo importante. O modelo precisa aprender padrões relevantes sem se tornar dependente demais do conjunto de treino (*overfitting*).

### 🔹 4. Avaliação
Utilizar métricas como acurácia permite medir a performance, mas é fundamental entender que métricas sozinhas não explicam tudo — principalmente em problemas desbalanceados.

### ✅ Conclusão
Cada etapa do ciclo de vida de um modelo é essencial. Um bom modelo não nasce de um algoritmo poderoso, mas sim de dados bem preparados, decisões conscientes e avaliações criteriosas. A prática e a iteração contínua são os caminhos para alcançar bons resultados.


## Exercício 11 Aprofundamento
Baseado nos exercícios anteriores, crie um novo modelo de classificação binário, agora para classificar os pinguins da classe Gentoo versus as outras espécies. O desempenho deste modelo é pior ou melhor do que o primeiro modelo criado?

---

```python

df['is_gentoo'] = (df['species'] == 'Gentoo').astype(int)
X_gentoo = df[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']]
y_gentoo = df['is_gentoo']

X_train_g, X_val_g, y_train_g, y_val_g = train_test_split(
    X_gentoo, y_gentoo, test_size=0.2, stratify=y_gentoo, random_state=42
)

modelo_gentoo = KNeighborsClassifier(n_neighbors=3)
modelo_gentoo.fit(X_train_g, y_train_g)

plt.figure(figsize=(8, 6))
plt.scatter(X_train_g['flipper_length_mm'], X_train_g['body_mass_g'], c=y_train_g, cmap='coolwarm', edgecolor='k')
plt.xlabel('Tamanho da Nadadeira (mm)')
plt.ylabel('Massa Corporal (g)')
plt.title('Gentoo vs Outros (Conjunto de Treinamento)')
plt.grid(True)
plt.show()
```

```python

y_pred_g = modelo_gentoo.predict(X_val_g)
acc_g = accuracy_score(y_val_g, y_pred_g)
print(f"Acurácia (Gentoo vs Outros): {acc_g:.2%}")


ConfusionMatrixDisplay.from_predictions(
    y_val_g, y_pred_g, display_labels=['Outros', 'Gentoo'], cmap='Greens'
)
plt.title('Matriz de Confusão - KNN (Gentoo)')
plt.show()
```

### 🧪 Classificação Binária: Gentoo vs Outras Espécies

- O modelo KNN foi treinado com `k=3` e os dados foram balanceados via `stratify`.
- A visualização com scatter plot ajuda a entender a separação das classes no espaço das features.
- A matriz de confusão mostra o desempenho real do modelo na validação.


## Exercício 12 Compreensão
Seria possível criar um modelo capaz de diferenciar as três espécies de pinguim? Escreva sua resposta destacando quais as diferenças seriam observadas na análise? 

---

### É possível criar um modelo que diferencie as três espécies de pinguim?

**Sim, é possível.**  
Ao invés de realizar uma classificação binária (como nos exercícios anteriores), podemos construir um modelo de **classificação multiclasse**, onde o `target` será a própria coluna `species` com três classes:  
- Adelie  
- Gentoo  
- Chinstrap

Modelos como **K-Nearest Neighbors**, **Random Forest**, **Logistic Regression (multinomial)** e redes neurais são capazes de lidar com múltiplas classes.

---

### Diferenças observadas na análise multiclasse:

- **Separação mais complexa:**  
  O modelo precisa aprender limites de decisão que separam *três grupos*, o que exige maior qualidade nos dados e pode reduzir um pouco a acurácia geral.
  
- **Avaliação com novas métricas:**  
  Métricas como *matriz de confusão*, *precisão por classe* e *F1-score macro* são mais indicadas do que apenas a acurácia.

- **Importância da visualização:**  
  Gráficos de dispersão ajudam a entender como as classes se sobrepõem ou se distinguem com base nas features (por exemplo, Gentoo costuma ter massa corporal maior, Adelie tem bico menor, etc).

---

### Conclusão

Sim, é viável — e comum — criar modelos multiclasse. A principal diferença está na complexidade de avaliação e separação das classes, o que torna o pré-processamento, a seleção de features e a análise exploratória ainda mais relevantes.
